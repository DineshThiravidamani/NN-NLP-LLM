```python
import pandas as pd
import tensorflow as tf
from tensorflow import keras
```

*Dataset*

```python
(x_train, y_train) ,( x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
x_train_flattend = x_train.reshape(len(x_train),28*28)
x_test_flattend = x_test.reshape(len(x_test),28*28)
```

```python
x_train_flattend
```

```python
x_train_flattend[1]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
x_train_flattend = StandardScaler().fit_transform(x_train_flattend)
x_train_flattend = StandardScaler().fit_transform(x_test_flattend)
```

```python
x_train_flattend[0]
```

```python
model = keras.Sequential([keras.layers.Dense(10,input_shape=(784,),activation='sigmoid')])
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model.fit(x_train_flattend,y_train,epochs=6)
```

```python
model.evaluate(x_test_flattend,y_test)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.matshow(x_test[1])
```

```python
y_predicted=model.predict(x_test_flattend)
```

```python
import numpy as np
```

```python
np.argmax(y_predicted[1])
```

```python
y_predicted_labels=[]
for i in y_predicted:
  np.argmax(i)
  y_predicted_labels.append(np.argmax(i))
```

```python
y_predicted_labels[:5]
```

```python
forhm=tf.math.confusion_matrix(y_test,y_predicted_labels)
forhm
```

```python
import seaborn as sns
sns.heatmap(forhm,annot=True,fmt='d')
```
