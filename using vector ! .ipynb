```python
import pandas as pd
```

```python
df = pd.read_csv('/content/Fake_Real_Data.csv')
```

```python
df.head()
```

```python
df['label_num'] = df.label.map({'Fake':0,'Real':1})
```

```python
df.head()
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
df['vector_text'] = df['Text'].apply(lambda x : nlp(x).vector)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.vector_text.values, df.label_num.values, test_size=0.2, random_state=43)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)

import numpy as np
X_train = np.stack(X_train)
X_test = np.stack(X_test)

knn.fit(X_train, y_train)

predict = knn.predict(X_test)

import sklearn.metrics as metrics
print(metrics.classification_report(y_test, predict))
```
