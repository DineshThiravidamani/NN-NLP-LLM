```python
import pandas as pd

df = pd.read_json('news_dataset.json')
```

```python
min_sample = 1381
df_bussiness = df[df.category == 'BUSINESS'].sample(min_sample,random_state=42)
df_sport = df[df.category == 'SPORTS'].sample(min_sample,random_state=42)
df_science = df[df.category == 'SCIENCE'].sample(min_sample,random_state=42)
df_crime = df[df.category == 'CRIME'].sample(min_sample,random_state=42)

df_balanced = pd.concat([df_bussiness,df_sport,df_science,df_crime],axis=0)
```

```python
df_balanced['category_num'] = df_balanced.category.map({'BUSINESS':0,'SPORTS':1,'SCIENCE':2,'CRIME':3})
```

```python
df.text
```

```python
df_balanced.category.value_counts()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_balanced.text, df_balanced.category_num, test_size=0.2, random_state=42,stratify=df_balanced.category_num)

from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer

model = Pipeline([('vectorizer', CountVectorizer()),('model', MultinomialNB())])
model.fit(X_train,y_train)

from sklearn.metrics import classification_report
predict=model.predict(X_test)
print(classification_report(y_test,predict))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_balanced.text, df_balanced.category_num, test_size=0.2, random_state=42,stratify=df_balanced.category_num)

from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer

model = Pipeline([('vectorizer', CountVectorizer(ngram_range=(1,2))),('model', MultinomialNB())])
model.fit(X_train,y_train)

from sklearn.metrics import classification_report
predict=model.predict(X_test)
print(classification_report(y_test,predict))
```
