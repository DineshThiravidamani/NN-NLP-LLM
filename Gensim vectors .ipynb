```python
import gensim.downloader as api
```

```python
wv = api.load('word2vec-google-news-300')
```

```python
import pandas as pd

df =pd.read_csv('/content/Fake_Real_Data.csv')
```

```python
df.head()
```

```python
df['label_num']=df['label'].map({'Real':1,'Fake':0})
```

```python
df.head()
```

```python
import spacy
nlp=spacy.load('en_core_web_sm')
```

```python
def preprocessing_and_vectorize(text):
  doc=nlp(text)

  filtered_text = []
  for token in doc:
    if token.is_stop or token.is_punct:
      continue
    filtered_text.append(token.lemma_)

  return wv.get_mean_vector(filtered_text)
```

```python
df['vectors'] = df['Text'].apply(lambda text : preprocessing_and_vectorize(text))
```
